```python
def repeat(n):
    def decorator_with_wrapper(func):
        def wrapper(*args,**kargs):
            for _ in range(n):
                func(*args,**kargs)
        return wrapper 
    return decorator_with_wrapper 

@repeat(3)
def func(str):
    print(f'{str}')
```
