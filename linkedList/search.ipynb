```python
class Node:
    def __init__(self,value):
        self.data=value
        self.next=None 
```

```python
def create_linked_list(arr:list):
    tail=None
    head=None
    for i in arr:
        new_Node=Node(i)
        if head==None:
            head=new_Node
            tail=new_Node
        else:
            tail.next=new_Node
            tail=new_Node
    return head
            
```

```python
def print_linked_list(head:Node):
    temp=head 
    while temp!=None:
        print(temp.data,end='->')
        temp=temp.next
```

```python
link=create_linked_list([2,3,4,5,6,7,8])
print_linked_list(link)
```

# search in LL by value 

```python
def search_value(head,value):
    index=0
    if head==None:
        return None 
    if head.data==value:
        return index
    temp=head 
    while temp!=None:
        if temp.data==value:
            return index
        index+=1
        temp=temp.next 
    return 'Not found'
```

```python
link=create_linked_list([2,3,4,5,6,7,8])
search_value(link,9)
```

```python
def search_value_recursive(head,value,index=0):
    if head==None:
        return None 
    if head.data==value:
        return index
     
    return search_value_recursive(head.next,value,index+1)
```

```python
link=create_linked_list([2,3,4,5,6,7,8])
search_value_recursive(link,4)
```

# search by index 

```python
def search_by_index(head,index):
    if head==None :
        return None 
    count=0
    temp=head
    while temp!=None and count<=index:
        if count==index:
            return temp.data 
        temp=temp.next
        count+=1 
    return -1 
```

```python
def search_by_index_recursive(head,index):
    if head==None:
        return None 
    if index==0:
        return head.data 
    return search_by_index_recursive(head.next,index-1)
```

```python
link=create_linked_list([2,3,4,5,6,7,8])
search_by_index_recursive(link,7)
```
