```python
# given a linked list - get the node that is in the middle
```

```python
class Node:
    def __init__(self,value):
        self.data=value
        self.next=None 

def create_linked_list(arr:list):
    tail=None
    head=None
    for i in arr:
        new_Node=Node(i)
        if head==None:
            head=new_Node
            tail=new_Node
        else:
            tail.next=new_Node
            tail=new_Node
    return head

def print_linked_list(head:Node):
    temp=head 
    while temp!=None:
        print(temp.data,end='->')
        temp=temp.next

def length_LL(head):
    if head==None:
        return 0
    count=0
    temp=head 
    while temp!=None:
        count+=1
        temp=temp.next
    return count

link=create_linked_list([2,3,4,5,6])
print_linked_list(link)
length_LL(link)
```

```python
def get_middle_LL(head):
    if head is None or head.next is None:
        return head
    ln=length_LL(head)
    temp=head 
    count=0
    while temp!=None and count<ln:
        if ln//2==count:
            return temp 
        temp=temp.next
        count+=1 
    return None 
```

```python
link1=create_linked_list([2,3,4,5,6,7])
link2=create_linked_list([2,3,4,5,6])
link3=create_linked_list([])
print(get_middle_LL(link1).data)
print(get_middle_LL(link2).data)
```

```python
# 2 pointer approach 
def get_middle_LL_two_pointer(head):
    if head is None or head.next is None:
        return head 
    first=head
    second=head 
    while second != None and second.next != None:
        first=first.next 
        second=second.next.next 
    return first
```

```python
link1=create_linked_list([2,3,4,5,6,6])
link2=create_linked_list([2,3,4,5,6])
print(get_middle_LL_two_pointer(link1).data)
print(get_middle_LL_two_pointer(link2).data)
```
