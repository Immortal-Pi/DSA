```python
class Node:
    def __init__(self,value):
        self.data=value
        self.next=None 

def create_linked_list(arr:list):
    head=None 
    tail=None
    for i in arr:
        newNode=Node(i)
        if head==None:
            head=newNode
            tail=newNode
        else:
            tail.next=newNode
            tail=newNode
    return head 

def print_linked_list(head:Node):
    temp=head 
    while temp!=None:
        print(temp.data,end='->')
        temp=temp.next
```

```python
link=create_linked_list([2,3,4,5,6,7,8])
print_linked_list(link)
```

# reverse a linked list 

```python
# 2 pointer approach 
def reverse_LL(head):
    if head==None or head.next==None:
        return head 
    prev=None 
    curr=head 
    while curr:
        temp=curr.next 
        curr.next=prev
        prev=curr 
        curr=temp 
         
    return prev 
```

```python
# recursive apprach 
def reverse_LL_recursive(head):
    if head==None or head.next==None:
        return head 
    
    reverse=reverse_LL_recursive(head.next)
    temp=reverse
    while temp.next!=None:
        temp=temp.next
    temp.next=head
    head.next=None
    return reverse 
```

```python
# recursive apprach 
def reverse_LL_recursive(head):
    if head==None or head.next==None:
        return head,head 
    
    reverse,tail=reverse_LL_recursive(head.next)
    tail.next=head
    head.next=None 
    tail=head
    return reverse,tail
```

```python
link=create_linked_list([2,3,4,5,6,7,8])
reverse,tail=reverse_LL_recursive(link)
# reverse=reverse_LL(link)
print_linked_list(reverse)
```

# merge sort a linked list 

```python
link=create_linked_list([3,2,10,5,8,12])
print_linked_list(link)
```

```python
def mid_LL(head):
    if head==None or head.next==None:
        return head
    first=head 
    second=head 
    while second!=None and second.next!=None:
        prev=first
        first=first.next 
        second=second.next.next
    prev.next=None 
    return head,first 

def merge(left,right):
    if left==None:
        return right 
    if right==None:
        return left 
    
    head=None
    tail=None
    while left!=None and right!=None:
        if left.data<right.data:
            if head==None:
                head=left 
                tail=left 
            else:
                tail.next=left
                tail=left 
            left=left.next 
        else:
            if head==None:
                head=right
                tail=right
            else:
                tail.next=right 
                tail=right 
            right=right.next 
    if left!=None:
        tail.next=left
    if right!=None:
        tail.next=right
    # print_linked_list(head)
    return head

def merge_sort(head):
    if head==None or head.next==None:
        return head
    left,right=mid_LL(head)
    left=merge_sort(left)
    right=merge_sort(right)
    # print_linked_list(left)
    return merge(left,right)
```

```python
sorted_list=merge_sort(link)
print_linked_list(sorted_list)
```
