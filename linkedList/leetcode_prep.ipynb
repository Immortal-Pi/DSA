```python
class Node:
    def __init__(self,value):
        self.data=value
        self.next=None 

def create_linked_list(arr:list):
    head=None 
    tail=None
    for i in arr:
        newNode=Node(i)
        if head==None:
            head=newNode
            tail=newNode
        else:
            tail.next=newNode
            tail=newNode
    return head 

def print_linked_list(head:Node):
    temp=head 
    while temp!=None:
        print(temp.data,end='->')
        temp=temp.next
```

```python
link=create_linked_list([2,3,4,5,6,7,8])
print_linked_list(link)
```
