```python
class Node:
    def __init__(self,value):
        self.data=value
        self.next=None 

def create_linked_list(arr:list):
    head=None 
    tail=None
    for i in arr:
        newNode=Node(i)
        if head==None:
            head=newNode
            tail=newNode
        else:
            tail.next=newNode
            tail=newNode
    return head 

def print_linked_list(head:Node):
    temp=head 
    while temp!=None:
        print(temp.data,end='->')
        temp=temp.next
```

```python
link=create_linked_list([2,3,4,5,6,7,8])
print_linked_list(link)
```

# reverse a linked list 

```python
# 2 pointer approach 
def reverse_LL(head):
    if head==None or head.next==None:
        return head 
    prev=None 
    curr=head 
    while curr:
        temp=curr.next 
        curr.next=prev
        prev=curr 
        curr=temp 
         
    return prev 
```

```python
# recursive apprach 
def reverse_LL_recursive(head):
    if head==None or head.next==None:
        return head 
    
    reverse=reverse_LL_recursive(head.next)
    temp=reverse
    while temp.next!=None:
        temp=temp.next
    temp.next=head
    head.next=None
    return reverse 
```

```python
# recursive apprach 
def reverse_LL_recursive(head):
    if head==None or head.next==None:
        return head,head 
    
    reverse,tail=reverse_LL_recursive(head.next)
    tail.next=head
    head.next=None 
    tail=head
    return reverse,tail
```

```python
def mid_LL(head):
    if head==None or head.next==None:
        return head
    first=head 
    second=head 
    while second!=None and second.next!=None:
        prev=first
        first=first.next 
        second=second.next.next
    prev.next=None 
    return head,first 

def merge(left,right):
    if left==None:
        return right 
    if right==None:
        return left 
    
    head=None
    tail=None
    while left!=None and right!=None:
        if left.data<right.data:
            if head==None:
                head=left 
                tail=left 
            else:
                tail.next=left
                tail=left 
            left=left.next 
        else:
            if head==None:
                head=right
                tail=right
            else:
                tail.next=right 
                tail=right 
            right=right.next 
    if left!=None:
        tail.next=left
    if right!=None:
        tail.next=right
    # print_linked_list(head)
    return head

def merge_sort(head):
    if head==None or head.next==None:
        return head
    left,right=mid_LL(head)
    left=merge_sort(left)
    right=merge_sort(right)
    # print_linked_list(left)
    return merge(left,right)
```

# double linked list 



```python
"""
# Definition for a Node.
class Node(object):
    def __init__(self, val, prev, next, child):
        self.val = val
        self.prev = prev
        self.next = next
        self.child = child
"""

class Solution(object):
    def flatten(self, head):
        """
        :type head: Node
        :rtype: Node
        """
        if not head:
            return head 

        temp=head
        while temp:
            if temp.child:
                child=temp.child
                second_half=temp.next 

                # connect the child 
                temp.next=child
                child.prev=temp
                temp.child=None 

                tail=child
                while tail.next:
                    tail=tail.next 

                # combine with second half
                tail.next=second_half
                if second_half:
                    second_half.prev=tail 
            temp=temp.next
        return head
            
```

# largest rectangle in Histogram


```python
def largestRectangleArea(heights):
    stack=[]
    max_area=0
    n=len(heights)
    for i in range(n+1):
        current_height=heights[i] if i<n else 0 
        while stack and heights[stack[-1]]>current_height:
            top=stack.pop()
            height=heights[top]
            width=i if not stack else i - stack[-1] -1 
            max_area=max(max_area,height*width)
        stack.append(i)
    return max_area
```
