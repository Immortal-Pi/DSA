# LinkedList


```python
# create a linkedList
class Node:
    def __init__(self,value):
        self.data=value 
        self.next=None
```

```python
first=Node(1)
second=Node(2)
third=Node(3)
```

```python
print(id(first),id(second),id(third))
```

```python
first.next=second
second.next=third
```

```python
print(first.next,second.next)
```

```python
print(int(0x0000029E786A6D50),int(0x0000029E786A7C50))
```

```python
head=first
del(first)
del(second)
del(third)
```

```python
# print all the elements in linkedlist
def print_linked_list(head:Node):
    temp=head
    while temp!=None:
        print(temp.data)
        temp=temp.next 
    temp=head 
    while temp!=None:
        print(temp.data)
        temp=temp.next 

print_linked_list(head)
```
