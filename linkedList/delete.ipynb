```python
class Node:
    def __init__(self,value):
        self.data=value
        self.next=None 
```

```python
def create_linked_list(arr:list):
    tail=None
    head=None
    for i in arr:
        new_Node=Node(i)
        if head==None:
            head=new_Node
            tail=new_Node
        else:
            tail.next=new_Node
            tail=new_Node
    return head
            
```

```python
def print_linked_list(head:Node):
    temp=head 
    while temp!=None:
        print(temp.data,end='->')
        temp=temp.next
```

```python
link=create_linked_list([2,3,4,5,6,7,8])
print_linked_list(link)
```

```python
# recursive delete 
def delete_tail_recursive(head):
    if head==None or head.next==None:
        return None  
    head.next=delete_tail_recursive(head.next)
    return head
```

```python
# delete ast gven index 
def delete_index(index,head):
    if head==None :
        print('nothing to delete')
        return None 
    if index==0:
        head=delete_head(head)
        return head
    count=0
    temp=head
    while temp!=None and count<index-1:
        temp=temp.next
        count+=1 
    if temp is None or temp.next is None:
        print('out of bound')
        return head
    temp.next=temp.next.next 
    return head         
```

```python
link=create_linked_list([2,3,4,5,6,7,8])
# print_linked_list(link)
link=delete_index(8,link)
print_linked_list(link)
```

```python
# delete head recrusively 
def delete_index_recursive(index,head):
    if head==None :
        print('nothing to delete')
        return None 
    if index==0:
        head=delete_head(head)
        return head
    head.next=delete_index_recursive(index-1,head.next)
    return head 
```

```python
link=create_linked_list([2,3,4,5,6,7,8])
# print_linked_list(link)
link=delete_index_recursive(6,link)
print_linked_list(link)
```

```python
def delete_value(value,head):
    if head==None:
        print(f'Nothing to delete')
        return None
    if head.data==value:
        head=head.next 

    temp=head 
    while temp.next!=None and temp.next!=None:
        if temp.next.data==value:
            temp.next=temp.next.next
        else:
            temp=temp.next 
    return head
```

```python
link=create_linked_list([2,3,2,3,4,5,6,7,8,8])
# print_linked_list(link)
link=delete_value(2,link)
print_linked_list(link)
```

```python
# delete by value recursively 
def delete_value_recursive(value,head):
    if head==None:
        print(f'Nothing to delete')
        return None
    if head.data==value:
        head=head.next
    

    head.next=delete_value_recursive(value,head.next)
    return head
```

```python
link=create_linked_list([8,7,8,8,9])
# print_linked_list(link)
link=delete_value_recursive(8,link)
print_linked_list(link)
```
