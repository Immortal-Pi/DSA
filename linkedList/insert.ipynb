```python
class Node:
    def __init__(self,value):
        self.data=value
        self.next=None 
```

```python
def create_linked_list(arr:list):
    tail=None
    head=None
    for i in arr:
        new_Node=Node(i)
        if head==None:
            head=new_Node
            tail=new_Node
        else:
            tail.next=new_Node
            tail=new_Node
    return head
            
```

```python
def print_linked_list(head:Node):
    temp=head 
    while temp!=None:
        print(temp.data,end='->')
        temp=temp.next
```

```python
link=create_linked_list([2,3,4,5,6,7,8])
print_linked_list(link)
```

# insert at head

```python
def insert_at_head(value,head:Node):
    new_Node=Node(value)
    if head==None:
        head=new_Node
    else:
        new_Node.next=head
        head=new_Node
    return head
```

```python
link=insert_at_head(44,link)
print_linked_list(link)
```

```python
def insert_at_tail(value,head:Node):
    new_Node=Node(value)
    if head==None:
        return new_Node
    
    temp=head
    while temp.next!=None:
        temp=temp.next
    temp.next=new_Node
    return head
```

```python
def insert_at_tail_recusive(value,head:Node): 
    if head==None:
        new_Node=Node(value)
        return new_Node
        
    head.next=insert_at_tail_recusive(value,head.next)
    return head
```

```python
link=create_linked_list([2,3,4,5])
link=insert_at_tail_recusive(12,link)
print_linked_list(link)
```

# insert at index 

```python
def insert_at_index(value, index, head):
    if index==0 or head==None:
        head=insert_at_head(value,head)
        return head
    new_Node=Node(value)
    count=0
    temp=head 
    while temp!=None and count<index-1:
        temp=temp.next
        count+=1
    new_Node.next=temp.next 
    temp.next=new_Node
    return head
```

```python
def insert_at_index_recursion(value,index,head):
    if index<0:
        raise IndexError("index cannot be negative")
    if index==0 :
        return insert_at_head(value,head)
    head.next=insert_at_index_recursion(value,index-1,head.next)
    return head
        
```

```python
link=create_linked_list([2,3,4,6,7,8])
# link=insert_at_index(12,1,link)
link=insert_at_index_recursion(12,6,link)
print_linked_list(link)
```
