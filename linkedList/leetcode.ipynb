```python
class Node:
    def __init__(self,value):
        self.data=value
        self.next=None 

def create_linked_list(arr:list):
    tail=None
    head=None
    for i in arr:
        new_Node=Node(i)
        if head==None:
            head=new_Node
            tail=new_Node
        else:
            tail.next=new_Node
            tail=new_Node
    return head

def print_linked_list(head:Node):
    temp=head 
    while temp!=None:
        print(temp.data,end='->')
        temp=temp.next

def length_LL(head):
    if head==None:
        return 0
    count=0
    temp=head 
    while temp!=None:
        count+=1
        temp=temp.next
    return count

link=create_linked_list([2,3,4,5,6])
print_linked_list(link)
length_LL(link)
```

# merge 2 sorted linked list

```python
def merge_sorted_LL(head1,head2):
    if head1 is None:
        return head2
    if head2 is None:
        return head1 
    finalHead=None
    finalTail=None 
  
    while head1!=None and head2!=None:
    
        if head1.data<head2.data:
            if finalHead==None:
                finalHead=head1
                finalTail=head1
            else:
                finalTail.next=head1
                finalTail=head1
            head1=head1.next
        else:
            if finalHead==None:
                finalHead=head2
                finalTail=head2
            else:
                finalTail.next=head2 
                finalTail=head2
            head2=head2.next
        if head1!=None:
            finalTail.next=head1
        if head2!=None:
            finalTail.next=head2 
        print(finalHead.data,finalTail.data)
    return finalHead
            
```

```python
list1=create_linked_list([2,4,6,6])
list2=create_linked_list([1,3,5,7])
merged=merge_sorted_LL(list1,list2)
print_linked_list(merged)
```

# merge sort linked list 

```python
def get_mid_LL(head):
    if head==None or head.next==None:
        return head
    first=head
    second=head 
    while second!=None and second.next!=None:
        prev=first
        first=first.next 
        second=second.next.next
    
    prev.next=None
    return head,first

def merge(left,right):
    if right==None:
        return left 
    if left==None:
        return right
    newHead=None 
    newTail=None
    while left!=None and right!=None:
        if left.data<right.data:
            if newHead==None:
                newHead=left
                newTail=left 
            else:
                newTail.next=left
                newTail=left 
            left=left.next 
        else:
            if newHead==None:
                newHead=right
                newTail=right 
            else:
                newTail.next=right
                newTail=right 
            right=right.next 
    if left!=None:
        newTail.next=left 
    if right!=None:
        newTail.next=right
    return newHead


def merge_sorted_LL(head):
    if head==None or head.next==None:
        return head
    left,right=get_mid_LL(head)
    left=merge_sorted_LL(left)
    right=merge_sorted_LL(right)
    # print_linked_list(left)
    return merge(left,right)
    
```

```python
link=create_linked_list([3,2,10,5,8,12])

merged=merge_sorted_LL(link)
print_linked_list(merged)
```

```python
def reverse(head):
    if head==None or head.next==None:
        return head
    prev=None 
    temp=head 
    
    while temp!=None:
        fw=temp.next
        temp.next=prev
        prev=temp
        temp=fw
        
    return prev

```

```python
link=create_linked_list([3,2,10,5,8,12])

merged=reverse(link)
print_linked_list(merged)
```
