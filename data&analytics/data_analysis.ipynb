```python
data.rename(columns={'Owner_Company':'Owner'},inplace=True)
data.head()
```

```python
# change datatypes 
data['Rounded_Estimated_Capacity_MW']=data['Estimated_Capacity_MW'].astype(int)
data.head()
```

```python
data['Surrounding_Water_Stress_Tier'].unique()
```

```python
def change_root(type):
    if type=='Low':
        return -1 
    if type=='Medium':
        return 1 
    if type=='High':
        return 1
    
data['Req']=data['Surrounding_Water_Stress_Tier'].apply(change_root)
data 
    
```

# aggregation

```python
grouped_data=data.groupby(['Cooling_System_Type'])['Daily_Water_Usage_Gallons'].mean()
grouped_data
```

```python
grouped_data=data.groupby(['Cooling_System_Type'])['Daily_Water_Usage_Gallons'].agg(['min','mean','max'])
grouped_data
```

```python
# joins .merge on key
df1=pd.DataFrame({'Keys':['A','B','C'],'Value1':[1,2,3]})
df2=pd.DataFrame({'Keys':['D','B','C'],'Value1':[4,7,10]})

pd.merge(df1,df2,on='Keys',how='inner')
```
