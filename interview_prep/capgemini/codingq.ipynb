```python
from collections import defaultdict 

def groupAnagram(strs):
    groups=defaultdict(list)
    print(groups)
    for word in strs:
        key=''.join(sorted(word))
        groups[key].append(word)
    print(groups)
    return list(groups.values())
input=["eat","tea","tan","ate","nat","bat"]
groupAnagram(input)
```

```python
# top k 

def top_k(arr,num):
    all={}
    for i in arr:
        if i not in all.keys():
            all[i]=1
        else:
            all[i]+=1 

    return [k for k,v in all.items() if v>=num]
print(top_k([1,1,1,2,2,3], 2))
```

```python
# merge
def merge(arr):
    print(sorted(arr))
    merged=[arr[0]]
    for start,end in arr[1:]:
        last=merged[-1]
        if start<=last[1]:
            last[1]=max(last[1],end)
        else:
            merged.append([start,end])
    return merged


merge([[1,3],[2,6],[8,10],[15,18],[5,9]])
```

# copy and shallow copy 

```python
# copy 
a=[1,2,3,4,5,6]
b=a 
b.append(7)
b[0]=22
print(a,b)
```

```python
# shallow copy
a=[1,2,3,4,5,6]
b=a.copy() 
b.append(7)
b[0]=33
print(a,b) 
```

```python
# deep copy
a=[[1,2,3],[10,22,30]]
b=a.copy()
b[0].append(30)
print(a,b)
```

```python
from copy import deepcopy
a=[[1,2,3],[10,22,30]]
b=deepcopy(a)
b[0].append(30)
print(a,b)
```

```python
# 2 sums 
def two_sum(nums:list,target:int):
    seen={}
    all=[]
    for i,num in enumerate(nums):
        completion=target-num 
        if completion in seen:
            all.append([seen[completion],i])
        seen[num]=i
        print(seen,all) 
    return all 

nums = [1,8,2, 7, 11, 15]
print(two_sum(nums, 9)) 
```

```python
# group anagram 
def group_anagrams(strs:list):
    seen={}
    for s in strs:
        if ''.join(sorted(s)) in seen:
            seen[''.join(sorted(s))].append(s)
        else:
            seen[''.join(sorted(s))]=[s]
    return seen.values()

strs = ["eat","tea","tan","ate","nat","bat"]
print(group_anagrams(strs))
```
