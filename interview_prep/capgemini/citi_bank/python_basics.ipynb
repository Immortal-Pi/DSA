```python
# frequency counting 
def freq(nums):
    seen={}
    for i in nums:
        seen[i]=seen.get(i,0)+1
    return [k for k,v in seen.items() if v//2]


freq([1,2,2,2,2,3,3,3,5])
```

```python
# 2 sum 
def two_sum(nums,target):
    seen={}
    all=[]
    for i,n in enumerate(nums):
        diff=target-n 
        if diff in seen:
            all.append([seen[diff],i])
        seen[n]=i 
    return all 




two_sum([1,2,3,4,5,6,7,8,8,9],8)
```

```python
# group anagrams 

def group_anagrams(words)->list[list]:
    dics={}
    for word in words:
        k1=''.join(sorted(word))
        if k1 in dics:
            dics[k1].append(word)
        else:
            dics[k1]=[word]
    return [v for k,v in dics.items()]



print(group_anagrams(["eat", "tea", "tan", "ate", "nat", "bat"]))
```
