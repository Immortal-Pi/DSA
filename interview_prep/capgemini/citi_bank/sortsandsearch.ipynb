```python
# linear search
def linear_search(arr,num):
    for i in range(len(arr)):
        if arr[i]==num:
            return i 
    return -1 
```

```python
# binary search
def binary_search(arr,num):
    start=0
    end=len(arr)-1
    while start<=end:
        mid=(start+end)//2
        if arr[mid]==num:
            return mid 
        elif arr[mid]>num:
            end=mid-1
        else:
            start=mid+1
    return -1 
```

```python
arr=sorted([99,22,44,1,2,7,3,4,0])
print(arr)
binary_search(arr,2)
```
