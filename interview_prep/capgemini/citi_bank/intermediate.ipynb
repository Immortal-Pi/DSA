```python
flowers=["Rose","Lily","Tulip","Orchid","Lily","Rose"]

def max_distance(flowers):
    values={}
    for i,flower in enumerate(flowers):
        if flower not in values:
            values[flower]=-i
        else:
            values[flower]+=i
    
    sort_values=list(dict(sorted(values.items(),key=lambda x:x[1],reverse=True)).values())[0]
    print(sort_values)
```

```python
flowers=["Jasmine","Daisy","Jasmine","Marigold","Daisy"]
max_distance(flowers)
```

```python
def working(count):
    def functio(func):
        def wrapper(*args,**kwargs):
            for num in range(count):
                func(*args,**kwargs)
            
        return wrapper
    return functio 

@working(5)
def say_hi(name):
    print(f'Hi! {name}')

say_hi('amruth')
```

```python
def make_multipliers():
    multipliers = []
    for i in range(3):
        def multiply(x,i=i):
            return x * i
        multipliers.append(multiply)
    return multipliers

fns = make_multipliers()
print([f(10) for f in fns])
```

```python
def safe_divide(a, b):
    try:
        return a / b
    except:
        return None

def process_batch(pairs):
    results = []
    for a, b in pairs:
        results.append(safe_divide(a, b))
    return results

print(process_batch([(10, 2), (5, 0), (8, "x")]))
```

```python
def remove_short_words(words):
    for word in words:
        if len(word) < 4:
            words.remove(word)
    return words

print(remove_short_words(["cat", "dog", "elephant", "ox", "bear"]))
```
