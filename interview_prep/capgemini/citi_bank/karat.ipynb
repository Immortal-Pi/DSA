```python
def group_order_totals(orders, totals=None):
    totals={}
    for order in orders:
        print(order)
        customer = order["customer"]
        if customer not in totals:
            totals[customer] = 0
        totals[customer] += order["amount"]
    return totals


def process_batches(batches):
    results = []
    for batch in batches:
        results.append(group_order_totals(batch))
    return results


orders_batch_1 = [
    {"customer": "alice", "amount": 100},
    {"customer": "bob", "amount": 50},
]

orders_batch_2 = [
    {"customer": "alice", "amount": 75},
    {"customer": "amruth", "amount": 65},

]

print(process_batches([orders_batch_1, orders_batch_2]))
```
