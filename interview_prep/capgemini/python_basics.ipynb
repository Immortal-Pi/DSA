# decorators 

```python
def instructions():
    def fun_args(func):
        def wrapper(*args,**kwargs):
            print(f'started before function')
            return func(*args,**kwargs)
            print(f'started after function')
        return wrapper
    return fun_args

@instructions()
def add1(*args):
    return sum(args)

a=[1,2,3,4,5]
print(add1(1,2,3,4))
```

```python
def instructions(num):
    for i in range(0,num):
        print(f'checking')
    def func(func):
        def wrapper(*args,**kwargs):
            return func(*args,**kwargs) 
        return wrapper
    return func

@instructions(3)
def add1(*args):
    return sum(args)

add1(1,2,3,4)
```

```python
# how generators help in memmory 
import re 
def process_text(input_str:str):
    if re.search(r'amruth',input_str):
        return True 
    else:
        return False
    
def read_logs(file_path):
    with open(file_path) as f:
        for line in f:
            yield line 

for log in read_logs('big.txt'):
    if process_text(log):
        print(log)
    else:
        print('NA')
```

```python
# context manager 
def read_book(path):
    with open(path) as f:
       for line in f:
           yield line 

c=read_book('big.txt')
```
