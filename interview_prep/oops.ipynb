# abstract method 
- every subclass must implement it 

```python
from abc import ABC, abstractmethod
class payment(ABC):

    @abstractmethod
    def process_payment(self,amount):
        self.amount=amount
        print(f'amount:{amount}') 


class creditcard(payment):

    def __init__(self):
        super().__init__() 
    
    def process_payment(self, amount):
        print(f'processing credit card payment')
        return super().process_payment(amount)
    
class debitcard(payment):

    def __init__(self):
        super().__init__() 
    
    def process_payment(self, amount):
        print(f'processing debit card payment')
        return super().process_payment(amount)



d=[creditcard(),debitcard()]
for p in d:
    p.process_payment(100)
```
