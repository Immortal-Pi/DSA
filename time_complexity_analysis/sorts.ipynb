# time complexity 

```python
import time 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# merge sort 
def merge_sort(arr:list)->list:
    n=len(arr)
    if n<=1:
        return arr 
    mid=n//2
    left=merge_sort(arr[:mid])
    right=merge_sort(arr[mid:])
    return merge(left,right)

def merge(left,right):
    i=j=0
    arr=[]
    while i<len(left) and j<len(right):
        if left[i]<right[j]:
            arr.append(left[i])
            i+=1
        else:
            arr.append(right[j])
            j+=1
    arr.extend(left[i:])
    arr.extend(right[j:])
    return arr 


```

```python
array=[10,-2,33,16,12,99,-3,-7]
input_size=[10**i for i in range(1,7)]
merge_sort_time=[]

for size in input_size:
    arr=np.arange(size,0,-1)
    # print(arr)
    start_time=time.time_ns()
    merge_sort(arr)
    end_time=time.time_ns()
    merge_sort_time.append(end_time-start_time)
    print(f'time taken to sort an array of size {size}:{end_time-start_time} nanoseconds')
```

```python
plt.plot(input_size,merge_sort_time)
plt.xlabel('Input size (n)')
plt.ylabel('Time Taken (nanoseconds)')
plt.title('Time taken complexity merge_Sort')
plt.grid()
plt.show()
```

```python

def bubble_sort(arr):
    for i in range(len(arr)):
        for j in range(len(arr)-i-1):
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
    return arr 
# bubble_sort(array)
```

```python
input_size=[10**i for i in range(1,4)]
bubble_sort_time=[]
for size in input_size:
    arr=np.arange(size,0,-1)
    # print(arr)
    start_time=time.time_ns()
    bubble_sort(arr)
    end_time=time.time_ns()
    bubble_sort_time.append(end_time-start_time)
    print(f'time taken to sort an array of size {size}:{end_time-start_time} nanoseconds')
```

```python
plt.plot(input_size,bubble_sort_time)
plt.xlabel('Input size (n)')
plt.ylabel('Time Taken (nanoseconds)')
plt.title('Time taken complexity merge_Sort')
plt.grid()
plt.show()
```

```python
# insertion sort time complexity 
def insertion_sort(arr):
    for i in range(0,len(arr)):
        current_element=arr[i]
        j=i-1 
        while j>=0 and arr[j]>current_element:
            arr[j+1]=arr[j]
            j-=1 
        arr[j+1]=current_element
    return arr 
        

array=[10,-2,33,16,12,99,-3,-7]
insertion_sort(array)
```

```python
# selection sort 
array=[10,-2,33,16,12,99,-3,-7]
def selection_sort(arr):
    for i in range(0,len(arr)):
        least=i
        for j in range(i+1,len(arr)):
            if arr[j]<arr[least]:
                least=j
        if least!=i:
            arr[i],arr[least]=arr[least],arr[i]
    return arr
selection_sort(array)
```

```python
input_size=[10**i for i in range(1,4)]
insertion_sort_time=[]
for size in input_size:
    arr=np.arange(size,0,-1)
    # print(arr)
    start_time=time.time_ns()
    selection_sort(arr)
    end_time=time.time_ns()
    insertion_sort_time.append(end_time-start_time)
    print(f'time taken to sort an array of size {size}:{end_time-start_time} nanoseconds')
```

```python
# factorial 
def factorial(n:int)->int:
    if n<=1:
        return 1 
    return n* factorial(n-1)

factorial(6)
```

```python
# binary search 
def binary_search(arr:list,element):
    # if len(arr)<=1:
    #     return arr
    start=0
    end=len(arr)-1
    while start<=end:
        mid=(start+end)//2
        if arr[mid]==element:
            return mid 
        elif element<arr[mid]:
            end=mid-1 
        else:
            start=mid+1 
    return -1 
array=[-7, -3, -2, 10, 12, 16, 33, 99]
binary_search(array,99)
```
