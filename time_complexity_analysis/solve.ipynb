```python
input=["eat","tea","tan","ate","nat","bat"]
```

```python
all=[]
indexes_done=[]
for i in range(len(input)):
    if i not in indexes_done:
        indi=[input[i]]
    else:
        continue
    indexes_done.append(i)
    for j in range(i,len(input)):
        if sorted(input[i])==sorted(input[j]) and j not in indexes_done:
            indexes_done.append(j)
            indi.append(input[j])
    all.append(indi)
print(all)
```

```python
all={}

for i in range(len(input)):
    # print(tuple(sorted(input[i])))
    k=tuple(sorted(input[i]))
    if k not in all:
        all[k]=[input[i]]
    elif k in all.keys():
        all[k].append(input[i])
print(all)
```
