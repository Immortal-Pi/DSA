```python
import random
import copy 
# ---------- Test harness ----------
def run_tests(sort_func, name):
    test_cases = [
        ("Empty", []),
        ("Single element", [5]),
        ("Two sorted", [1, 2]),
        ("Two reversed", [2, 1]),
        ("Already sorted", [1, 2, 3, 4, 5]),
        ("Reverse sorted", [5, 4, 3, 2, 1]),
        ("All duplicates", [7, 7, 7, 7]),
        ("Some duplicates", [3, 1, 3, 2, 1, 3]),
        ("Negatives", [-3, -1, -5, 0, 2]),
        ("Mixed with zero", [0, -1, 1, -2, 2]),
        ("Nearly sorted", [1, 2, 4, 3, 5, 6]),
        ("Single spike", [1, 1, 1, 9, 1, 1]),
        ("Large random", random.sample(range(-500, 500), 100)),
    ]
    print(f"\n=== Testing {name} ===")
    passed = failed = 0
    
    for case_name, arr in test_cases:
        original = copy.deepcopy(arr)
        expected = sorted(original)        # Python's built-in is our oracle
        result = sort_func(copy.deepcopy(original))
        
        if result == expected:
            print(f"  ✅ {case_name}")
            passed += 1
        else:
            print(f"  ❌ {case_name}")
            print(f"     Input:    {original}")
            print(f"     Expected: {expected}")
            print(f"     Got:      {result}")
            failed += 1
    
    print(f"  → {passed} passed, {failed} failed")
```

```python
# bubble sort 
def bubble_sort(input_list):
    for i in range(len(input_list)):
        for j in range(len(input_list)-1):
            if input_list[j]>input_list[j+1]:
                input_list[j],input_list[j+1]=input_list[j+1],input_list[j]
    return input_list
```

```python
# selection sort 
def selection_sort(input_list):
    for i in range(len(input_list)):
        min_index=i 
        for j in range(i,len(input_list)):
            if input_list[j]<input_list[min_index]:
                min_index=j
        if min_index!=i: 
            input_list[i],input_list[min_index]=input_list[min_index],input_list[i]
    return input_list
```

```python
# insertion sort 
def insertion_sort(input_list):
    for i in range(1,len(input_list)):
        current_element=input_list[i]
        current_index=i-1
        for j in range(current_index,-1,-1):
            # print('b',input_list,j,current_element)
            if input_list[j]>current_element:
                input_list[j+1]=input_list[j]
                current_index-=1
        input_list[current_index+1]=current_element
        # print('a',input_list,j,current_element)
    return input_list

```

```python
# quick sort 
def quick_sort(input_list):
    if len(input_list)<=1:
        # print(input_list)
        return input_list
    pivot=input_list[0]
    left=[]
    right=[]
    # print(left,pivot,right)
    for i in range(1,len(input_list)):
        if input_list[i]<=pivot:
            left.append(input_list[i])
        else:
            right.append(input_list[i])
    # print(f'left:{left}')
    # print(f'pivot:{pivot}')
    # print(f'right:{right}')
    return quick_sort(left) + [pivot] + quick_sort(right)

```
