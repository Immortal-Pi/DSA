```python
def checks(func):
    def wrapper(*args, **kwargs):
        print(f'before call')
        result=func(*args,**kwargs)
        print(f'after call')
        return result
    return wrapper



@checks 
def add(a:int, b:int)->float:
    return a+b 
```

```python
print(add(5,4))
```

```python
def repeat(n):
    def decorator(func):
        def wrapper(*args,**arwgs):
            for i in range(n):
                func(*args,**arwgs)
        return wrapper 
    return decorator 
```
