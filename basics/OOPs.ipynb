```python
class car:
    def __init__(self,model,year,make):
        self.__model=model
        self.__year=year
        self.__make=make

class truck:
    def __init__(self,wheels):
        self.__wheels=wheels 

class electric(car,truck):
    def __init__(self, model, year, make,wheels):
        car.__init__(self,model, year, make)
        truck.__init__(self,wheels)

    @property
    def model(self):
        return self.model 
    
    @model.setter 
    def set_model(self,model):
        self.__model=model

```

```python
tesla=electric('T1',2025,'tesla',4)
```

```python
def square(n):
    for i in range(n):
        yield i**2 
```
