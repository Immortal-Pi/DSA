```python
def subsequences(string:str)->list:
    if string=='':
        return ['']
    
    sub_seq=subsequences(string[1:])
   
    ans=[]
    ans.extend(sub_seq)
    for i in sub_seq:
        ans.append(string[0]+i)
    # print(string)
    return ans

subsequences('abcd')
    

```

```python
# print all the subsequence
def print_subsequences(string:str,taken_so_far)->list:
    if string=='' or len(string)==0:
        print(taken_so_far) 
        return 
    
    current_char=string[0]
    smaller_input=string[1:]
    print_subsequences(smaller_input,taken_so_far+current_char)
    print_subsequences(smaller_input,taken_so_far)    
    return 

print_subsequences('abc','')
```

```python
# permultations of a string
def print_permutation(string:str, taken_so_far:str)->list:
    final=[]
    if string=='' or len(string)==0:
        print(taken_so_far)
        return 
    first_char=string[0]
    sub_seq=string[1:]
    
    for i in range(0,len(taken_so_far)+1):
        print_permutation(sub_seq,taken_so_far[0:i]+first_char+taken_so_far[i:])

    return  


print_permutation('abc','')
```

```python
# permultations of a string
def get_permutation(string:str)->list:
    if string=='':
        return ['']
    
    char=string[0]
    smaller_seq=get_permutation(string[1:])
    all=[]
    for each in smaller_seq:
        for i in range(0,len(each)+1):
            all.append(each[0:i]+char+each[i:])
    return all 

    
get_permutation('abcd')
```

```python
# return all codes
# input : '1123' -> '1-a, 2-b, ... 26-z' 
# codes :  1 1 2 3  and 11  23 and 1 12 3 and 1 1  23 and 11 2  3 
def num_to_char(num:int):
    return chr(num+64)

def get_codes(s:str):
    if s=='':
        return ['']
    singleDigit=int(s[0:1])
    doubleDigit=int(s[0:2])
    all=[]
    wordWithoutsingleDigit=get_codes(s[1:])
    for word in wordWithoutsingleDigit:
        all.append(num_to_char(singleDigit)+word)
    if doubleDigit>=10 and doubleDigit<=26:
        wordWithoutdoubleDigit=get_codes(s[2:])
        for word in wordWithoutdoubleDigit:
            all.append(num_to_char(doubleDigit)+word)
    return all 
    
get_codes('1123')
```

```python
# tower of hanio 
def tower_of_hanoi(n,s,d,a,rods):
   
    if n==0:
        return  
    tower_of_hanoi(n-1,s,a,d,rods)
    disk=rods[s].pop()
    rods[d].append(disk)
    print(f"{s} ----> {d}   (disk {disk})")
    print("   ", rods)  
    tower_of_hanoi(n-1,a,d,s,rods)
n=3  
rods = {'s': list(range(n, 0, -1)), 'd': [], 'a': []}
print("start:", rods)
tower_of_hanoi(3,'s','d','a',rods)
```
