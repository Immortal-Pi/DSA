```python
class Queue:
    def __init__(self):
        self.__queue=[]


    def enqueue(self,item):
        if len(self.__queue)==0:
            self.__queue.append(item)

        else:
            self.__queue.append(item)

    def dequeue(self):
        if len(self.__queue)==0:
            print(f'Nothing to remove')
            return 
        else:
            item=self.__queue[0]
            self.__queue.pop(0)
            return item

    def __str__(self):
        for i in self.__queue:
            print(i,end='<-')
        return f'..'
    
    def front(self):
        if len(self.__queue)==0:
            print('queue empty')
        else:
            return self.__queue[0]
        
    def rear(self):
        if len(self.__queue)==0:
            print('queue empty')
        else:
            return self.__queue[-1] 

    def is_empty(self):
        if len(self.__queue)==0:
            return True
        else:
            return False

    def size(self):
        return len(self.__queue) 
        
```

```python
q1=Queue()
q1.enqueue(1)
q1.enqueue(2)
q1.enqueue(3)
q1.enqueue(4)
q1.dequeue()
q1.enqueue(4)
print(q1)
q1.dequeue()
q1.enqueue(44)
print(q1)
print(q1.front())
print(q1.rear())
print(q1.size())
```

# Implementation using Linked list 

```python
class Node:
    def __init__(self,item):
        self.val=item
        self.next=None

class QueueLL:
    def __init__(self):
        self.__queue=None
        self.__size=0
        self.__front=None
        self.__rear=None 

    def enqueue(self,item):
        newNode=Node(item)
        if self.__queue==None:
            self.__queue=newNode
            self.__front=self.__queue
            self.__rear=self.__queue
            self.__size+=1 
        else:
            self.__rear.next=newNode
            self.__rear=newNode
            self.__size+=1 

    def dequeue(self):
        if self.__size==0 or self.__queue==None:
            print(f'Queue empty')
        else:
            item=self.__front.val
            self.__queue=self.__queue.next
            self.__front=self.__queue
            self.__size-=1 
            return item
    
    def is_empty(self):
        if self.__size==0:
            return True
        else:
            return False 
        
    def get_front(self):
        if self.__size==0:
            return f'Empty'
        return self.__front.val
    
    def get_rear(self):
        if self.__size==0:
            return f'Empty'
        return self.__rear.val 
    
    def get_size(self):
        return self.__size
    
    def __str__(self):
        if self.__queue==None:
            print(f'empty')
        temp=self.__queue
        while temp:
            print(temp.val,end='<-')
            temp=temp.next 
        return f''
```
