```python
class StackUsingList:
    def __init__(self):
        self.__stack=[]

    def push(self,item):
        self.__stack.append(item)

    def pop(self):
        if len(self.__stack)>0:
            self.__stack.pop()
        else:
            print(f'empty list')

    def size(self):
        return len(self.__stack)

    def is_empty(self):
        if len(self.__stack)==0:
            return True
        else: 
            return False

    def top(self):
        return self.__stack[-1]
    
    def __str__(self):
        return f'stack:{self.__stack}'

```

```python
stack1=StackUsingList()
stack1.push(3)
stack1.push(2)
stack1.push(2)
stack1.push(2)
stack1.size()
```

```python
print(stack1)
print(stack1.is_empty())
print(stack1.top())
stack1.pop()
stack1.pop()
print(stack1)
```

# Linkedlist approach 

```python
class Node:
    def __init__(self,value):
        self.val=value 
        self.next=None

class Stack_LL:

    def __init__(self):
        self.__node=None
        self.__tail=None
        self.__size=0

    
    def push(self,item):
        if self.__node==None:
            self.__node=Node(item)
            self.__tail=self.__node 
            self.__size+=1
        else:
            newNode=Node(item)
            newNode.next=self.__node
            self.__node=newNode
            self.__size+=1
    


    def pop(self):
        if self.__size>0:
            self.__node=self.__node.next 
            self.__size-=1
        else:
            print(f'empty list')

    def size(self):
        return self.__size

    def is_empty(self):
        if self.__size==0:
            return True
        else: 
            return False

    def top(self):
        if self.__node is None or self.__size==0:
            return "stack is empty"
        return self.__node.val 
    
    def __str__(self):
        temp=self.__node
        while temp:
            print(f'{temp.val}',end='->')
            temp=temp.next
        print(f'tail: {self.__tail.val}')
        return f'end'
```

```python
s1=Stack_LL()
s1.push(1)
s1.push(2)
s1.push(3)
s1.push(10)
s1.push(33)
print(s1)
s1.pop()
s1.pop()
s1.pop()
s1.pop()
s1.pop()
s1.pop()
# print(s1.__tail)



print(s1)
```
