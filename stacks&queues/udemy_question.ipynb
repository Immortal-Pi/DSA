# Valid Panthesis

```python
def is_balanced(S):
    #Write down your code here
    stack=[]
    mapping={')':'(','}':'{',']':'['}
    for s in S:
        if s in mapping.values():
            stack.append(s)
        if s in mapping:
            if not stack or mapping[s]!=stack.pop():
                return False
    return not stack

```

```python
def removeDuplicates(s: str) -> str:
    #Write your code here
    stack=[]
    for i in s:
        if len(stack)>0 and i==stack[-1]:
            stack.pop()
        else:
            stack.append(i)
    # print(stack)
    return ''.join(stack)
```

# next smallest element

Input: arr = [10, 20, 30, 40]
Output: [-1, -1, -1, -1]
 
Input: arr = [1, 3, 2, 4]
Output: [-1, 2, -1, -1]
 
Input: arr = [4, 5, 2, 10, 8]
Output: [2, 2, -1, 8, -1]

```python
def next_smallest_element(arr):
    n=len(arr)-1
    ans=[]
    stack=[]
    while n>=0:
        if len(stack)==0:
            stack.append(arr[n])
            ans.append(-1)
        elif len(stack)>0 and arr[n]<stack[-1]:
            stack.pop()
            n+=1
        else:
            ans.append(stack[-1])
            stack.append(arr[n])
        n-=1
    return ans[::-1]
```

```python
print(next_smallest_element([10,20,30,40]))
print(next_smallest_element([1,3,2,4]))
print(next_smallest_element([4,5,2,10,8]))
```

Start at the 1st friend.

Count k friends in a clockwise direction, including the starting friend. The counting wraps around the circle.

The friend at the k-th position leaves the circle.

If more than one friend remains, continue from the friend immediately clockwise of the one who just left.

The last remaining friend is the winner.

Given the number of friends n and an integer k, return the winner of the game.



Input:

n (1 ≤ n ≤ 10^5): The number of friends.

k (1 ≤ k ≤ 10^5): The counting number.

# Largest rectangle in Histogram

```python
def get_largest_rectangle(heights):
    max_area=0
    stack=[]
    n=len(heights)
    for i in range(n+1):
        current_height=heights[i] if i<n else 0
        while stack and heights[stack[-1]]>current_height:
            top=stack.pop()
            height=heights[top]
            width=i if not stack else i-stack[-1]-1
            max_area=max(max_area,height*width)
        stack.append(i)
    return max_area
            


            
             
```

```python
get_largest_rectangle([2, 1, 5, 6, 2, 3])
```

```python
class MyStack(object):

    def __init__(self):
        self.stack=[]
        

    def push(self, x):
        """
        :type x: int
        :rtype: None
        """
        self.stack.append(x)

        

    def pop(self):
        """
        :rtype: int
        """
        if len(self.stack)>0:
            
            return self.stack.pop()
        else:
            return -1 
        

    def top(self):
        """
        :rtype: int
        """
        return self.stack[-1] if self.stack else -1 
        

    def empty(self):
        """
        :rtype: bool
        """
        return True if len(self.stack)>0 else False
        


# Your MyStack object will be instantiated and called as such:
# obj = MyStack()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.top()
# param_4 = obj.empty()
```
