```python
list1=[1,2,4,5]
target=9

def get_target_sum_pair(list,target):
    p1=0
    p2=len(list)-1
    all=[]
    while p1<p2:
        s=list1[p1]+list1[p2]
        if s==target:
            all.append([p1,p2])
            p1+=1
            p2-=1 
        elif s>target:
            p2-=1
        else:
            p1+=1 
    return all 

get_target_sum_pair(list1,target)
        
    
```
