```python
class Solution(object):
    def reverseVowels(self, s):
        """
        :type s: str
        :rtype: str
        """
        vowels=set('aeiouAEIOU')
        s_list=list(s)
        left,right=0,len(s_list)-1

        while left<right:
            if s_list[left] not in vowels:
                left+=1
            elif s_list[right] not in vowels:
                right-=1
            else:
                s_list[left],s_list[right]=s_list[right],s_list[left]
                left+=1
                right-=1
        return ''.join(s_list)


        
s=Solution()
s.reverseVowels("IceCreAm")
```

```python
class Solution(object):
    def reverseWords(self, s):
        """
        :type s: str
        :rtype: str
        """
        words=s.split(' ')
        # print(words)
        new_string=''
        for word in range(len(words)-1,-1,-1):
            new_string+=' '+words[word]
        # print(new_string)
        return new_string.strip()
s=Solution()
s.reverseWords("the sky is blue")
```

```python
class Solution(object):
    def reverseWords(self, s):
        """
        :type s: str
        :rtype: str
        """
        words=s.split()
        reverserd_words=words[::-1]
        # print(new_string)
        return ' '.join(reverserd_words)
```

```python
class Solution(object):
    def reverseWords(self, s):
        """
        :type s: str
        :rtype: str
        """
        ans=""
        req=[]
        k=list(map(str,s.split(" ")))
        for i in k:
            if i=="":
                pass
            else:
                req.append(i)
        req=req[::-1]
        for i in req:
            ans=ans+i+" "
        return ans[:len(ans)-1]
```

# search insert position 


```python
class Solution(object):
    def searchInsert(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        start=0
        end=len(nums)-1
        while start<=end:
            mid=(start+end)//2
            # print(mid)
            if target>nums[mid]:
                start=mid+1
            elif target<nums[mid]:
                end=mid-1
            else:
                return mid
        return start 
            

Solution().searchInsert(nums=[1,3,5,6,10,20,21,22],target=11)  
```
