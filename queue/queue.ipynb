```python
from collections import deque 
class RecentCounter(object):

    def __init__(self):
        self.q=deque()
        

    def ping(self, t):
        """
        :type t: int
        :rtype: int
        """
        self.q.append(t)
        
        while t-self.q[0]>3000:
            print(self.q[0])
            self.q.popleft()
        return len(self.q)
```

```python
obj=RecentCounter()
obj.ping(1)
```
